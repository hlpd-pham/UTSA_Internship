```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('FTE Salary.csv')
```

```python
df.drop(['Unnamed: 17','UnitID'],axis=1,inplace=True)
```

```python
df_col = ['Institutional Name',
          'professor 2016', 'associate 2016', 'assistant 2016',
          'professor 2015', 'associate 2015', 'assistant 2015',
          'professor 2014', 'associate 2014', 'assistant 2014',
          'professor 2013', 'associate 2013', 'assistant 2013',
          'professor 2012', 'associate 2012', 'assistant 2012']
```

```python
df.columns = df_col
```

```python
df.info()
```

```python
median = {'Institutional Name': 'Median'}
```

```python
df.head()
```

```python
for column in df[df['Institutional Name'] != 'The University of Texas at San Antonio'].drop('Institutional Name',axis=1):
    median[column] = df[column].median()
```

```python
median_df = pd.DataFrame([median])
median_df
```

```python
df = df.append(median_df)
```

```python
df
```

```python
dropped_df = df[(df['Institutional Name'] == 'The University of Texas at San Antonio') | (df['Institutional Name'] == 'Median')] 
```

```python
dropped_df
```

```python
year_2012 = dropped_df[['assistant 2012','associate 2012','professor 2012']]
year_2013 = dropped_df[['assistant 2013','associate 2013','professor 2013']]
year_2014 = dropped_df[['assistant 2014','associate 2014','professor 2014']]
year_2015 = dropped_df[['assistant 2015','associate 2015','professor 2015']]
year_2016 = dropped_df[['assistant 2016','associate 2016','professor 2016']]
```

```python
year_2012
```

```python
list_of_years = [year_2012, year_2013, year_2014, year_2015, year_2016]
```

```python
for year in list_of_years:
    for index, row in year.iterrows():
        row['average'] = np.mean(row)
```
