```python
import pandas as pd
import numpy as np
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('PeerData.csv')
```

```python
df['Institution Name'].unique()
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Unnamed: 12','UnitID'],axis=1,inplace=True)
```

```python
df_names = ['Institution Name','4 Year - 2016 (100%)', '6 Year - 2016 (150%)', 
            '4 Year - 2015 (100%)', '6 Year - 2015 (150%)','4 Year - 2014 (100%)', '6 Year - 2014 (150%)', 
            '4 Year - 2013 (100%)', '6 Year - 2013 (150%)','4 Year - 2012 (100%)', '6 Year - 2012 (150%)',]
```

```python
df.columns = df_names
```

```python
df.info()
```

```python
df
```

```python
peers = df[df['Institution Name'] != 'The University of Texas at San Antonio'].drop(['Institution Name'],axis=1)
```

```python
medians = {'Institution Name': 'Median'}
```

```python
for column in peers:
    medians[column] = peers[column].median()
```

```python
medians
```

```python
median_df = pd.DataFrame([medians])
```

```python
UTSA = df[df['Institution Name'] == 'The University of Texas at San Antonio']
```

```python
UTSA
```

```python
median_df = median_df.append(UTSA)
```

```python
median_df
```

```python
test_df = pd.melt(median_df, id_vars="Institution Name", var_name='Year - Type', value_name='Percent')
```

```python
test_df
```

```python
test_df = test_df.sort_values(['Institution Name'])
```

```python
test_df
```

```python
test_df['Median'] = test_df[test_df['Institution Name']=='Median']['Percent']
```

```python
test_df['The University of Texas at San Antonio'] = test_df[test_df['Institution Name']=='The University of Texas at San Antonio']['Percent']
```

```python
test_df
```

```python
median = test_df.drop(['Institution Name','Percent','The University of Texas at San Antonio'],axis=1)
```

```python
median.dropna(axis=0,inplace=True)
```

```python
UTSA = test_df.drop(['Institution Name','Percent','Median'],axis=1)
```

```python
UTSA.dropna(axis=0,inplace=True)
```

```python
median_df = pd.merge(median,UTSA,how='inner',on='Year - Type')
```

```python
median_df = median_df.sort_values('Year - Type')
```

```python
median_df.to_csv('Aspirational.csv')
```

```python
median_df
```
