```python
#Important python libraries for data manipulating
import numpy as np
import pandas as pd
```

```python
# Read excel files
total_research = pd.read_excel('total_research.xlsx',header=2)
federal_research = pd.read_excel('federal_research.xlsx',header=2)
endowment_asserts = pd.read_excel('endowment_assets.xlsx',header=2)
annual_giving = pd.read_excel('annual_giving.xlsx',header=2)
natl_academy = pd.read_excel('natl_academy.xlsx',header=2)
faculty_awards = pd.read_excel('faculty_awards.xlsx',header=2)
dr_awards = pd.read_excel('doctorates.xlsx',header=2)
postdocs = pd.read_excel('postdocs.xlsx',header=2)
```

```python
data = [endowment_asserts, annual_giving, natl_academy, faculty_awards, dr_awards,
        total_research, federal_research, postdocs]
```

```python
key_cols = ['Research Group','Control', 
    'Institutions Reporting Any\nFederal Research in Past Five Years\n(in Alphabetical Order)'
]
cols_total = key_cols.copy()
cols_fed = key_cols.copy()
cols_endowment = key_cols.copy()
cols_giving = key_cols.copy()
cols_nat_member = key_cols.copy()
cols_fac_award = key_cols.copy()
cols_doc_award = key_cols.copy()
cols_postdoc = key_cols.copy()
```

```python
for column in total_research.columns:
    if column not in key_cols:
        cols_total.append(column + ' - TR')
total_research.columns = cols_total
```

```python
for column in federal_research.columns:
    if column not in key_cols:
        cols_fed.append(column + ' - FR')
federal_research.columns = cols_fed
```

```python
for column in endowment_asserts.columns:
    if column not in key_cols:
        cols_endowment.append(column + ' - EA')
endowment_asserts.columns = cols_endowment
```

```python
for column in annual_giving.columns:
    if column not in key_cols:
        cols_giving.append(column + ' - AG')
annual_giving.columns = cols_giving
```

```python
for column in natl_academy.columns:
    if column not in key_cols:
        cols_nat_member.append(column + ' - NM')
natl_academy.columns = cols_nat_member
```

```python
for column in faculty_awards.columns:
    if column not in key_cols:
        cols_fac_award.append(column + ' - FA')
faculty_awards.columns = cols_fac_award
```

```python
for column in dr_awards.columns:
    if column not in key_cols:
        cols_doc_award.append(column + ' - DA')
dr_awards.columns = cols_doc_award
```

```python
for column in postdocs.columns:
    if column not in key_cols:
        cols_postdoc.append(column + ' - PA')
postdocs.columns = cols_postdoc
```

```python
merge_cols = ['Institutions Reporting Any\nFederal Research in Past Five Years\n(in Alphabetical Order)',
        'Research Group', 'Control'
]
new_df = pd.merge(total_research,federal_research,on=key_cols)
new_df = pd.merge(new_df,endowment_asserts,on=key_cols)
new_df = pd.merge(new_df,annual_giving,on=key_cols)
new_df = pd.merge(new_df,natl_academy,on=key_cols)
new_df = pd.merge(new_df,faculty_awards,on=key_cols)
new_df = pd.merge(new_df,dr_awards,on=key_cols)
new_df = pd.merge(new_df,postdocs,on=key_cols)
```

```python
new_df.to_excel('combined dataset.xlsx')
```
